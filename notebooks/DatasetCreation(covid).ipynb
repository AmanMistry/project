```
import pandas as pd
import os
import shutil
import glob
```

```
!git clone https://github.com/agchung/Actualmed-COVID-chestxray-dataset.git
```

```
files_path='Actualmed-COVID-chestxray-dataset/metadata.csv'
images_path='Actualmed-COVID-chestxray-dataset/images'
```

```
df=pd.read_csv(files_path)
df.head()
```

```
Target_dir="Dataset/covid"
cnt=0
for (i,row) in df.iterrows():
  if row['finding']=='COVID-19' and row['view']=='PA':
    filename=row['imagename']
    image_path=os.path.join(images_path,filename)
    image_copy_path=os.path.join(Target_dir,filename)
    shutil.copy2(image_path,image_copy_path)
    cnt+=1
print(cnt," images are moved")
```

```
##now second dataset
!git clone https://github.com/ieee8023/covid-chestxray-dataset.git 
```

```
files_path='covid-chestxray-dataset/metadata.csv'
images_path='covid-chestxray-dataset/images'
```

```
df=pd.read_csv(files_path)
df.head()
```

```
cnt=0
for (i,row) in df.iterrows():
  if row['finding']=='Pneumonia/Viral/COVID-19' and row['view']=='PA':
    filename=row['filename']
    image_path=os.path.join(images_path,filename)
    image_copy_path=os.path.join(Target_dir,filename)
    shutil.copy2(image_path,image_copy_path)
    cnt+=1
print(cnt," images are moved")
```

```
##now another 2 dataset have not metadata.csv file so we have to done manually finding
```

```
from zipfile import ZipFile
fname='/content/New folder (3).zip'
with ZipFile(fname,'r') as zip:
  zip.extractall()
  print('done')
```

```
Source_dir='/content/New folder (3)/New folder (2)'
new_files=os.listdir(Source_dir)
old_files=os.listdir(Target_dir)
cnt=0
for f in new_files:
  image_path=os.path.join(Source_dir,f)
  image_copy_path=os.path.join(Target_dir,f)
  if(f not in old_files):
    shutil.copy2(image_path,image_copy_path)
    cnt+=1
print(cnt," images are moved")
```
